```python
# Import convention
from sklearn.cluster import KMeans

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load
dataRaw = np.loadtxt('exampleData.csv', delimiter=',', dtype='object')
header = dataRaw[0,:]
data = dataRaw[1:,:2]
data = np.vstack(data.astype(np.float32))
```

```python
# Plot
fig, ax = plt.subplots()
ax.scatter(data[:,0],data[:,1])
ax.set_title('Rodent data')
ax.set_xlabel(header[0])
ax.set_ylabel(header[1])
```

```python
#n_init is selected to be 15 is not determine by anything
#but randomly selected
#you are selecting a number of time you want
#the kmeans algorithm to run with different
#cendroid until it finds the best cendriod for all the data points

#the number of clusters are decided by your domain knowledge
kmeans = KMeans(n_clusters=3, n_init=15)
kmeans.fit(data)
print("WCSS: ", kmeans.inertia_)
print("Iternations until converged: ", kmeans.n_iter_)
print("Final centroids: ")
print(kmeans.cluster_centers_)
print("Cluster assignments ")
print(kmeans.labels_)
```

```python
fig, ax = plt.subplots()
idxs = np.unique(kmeans.labels_)

for i in idxs:
    #Points of cluster
    points = data[kmeans.labels_==i,:]
    #Plot points
    plt.scatter(points[:,0], points[:,1])
    #Plot centroids
    plt.scatter(kmeans.cluster_centers_[i][0], kmeans.cluster_centers_[i][1], s=100, c='red')

#Aesthetics
ax.set_title('K-Means clustering')
ax.set_xlabel(header[0])
ax.set_ylabel(header[1])
```
